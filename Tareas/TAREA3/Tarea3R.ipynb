```r
ordenar_col_df <- function(col){
    col <- col[complete.cases(col)]
    col <- sort(col,decreasing=TRUE) # Orden directo
    return (col)
}
```

```r
path = "weight-height.csv"
y = read.csv(path,header = TRUE) # lee un archivo csv
datos = y[["Height"]]
orden = ordenar_col_df(datos)
head(orden)
```

```r
contar_palabras <- function(datos){
    count=0
    lista_palabras <- strsplit(datos, split=" ")
    lista_palabras
    lon = matrix(lista_palabras)
    return(lon)
}
```

```r
path = "SMS_DATA.csv"
y= read.csv(path,header = TRUE) # lee un archivo csv
datos = toString(y[["Texto"]])
palabras_total = contar_palabras(datos)
print(palabras_total)
```

```r
palabras_dis <- function(datos){
    palabras_distintas=vector()
    lista_palabras <- strsplit(datos, split=" ")
    for(palabra in lista_palabras){
      D2 = setdiff(palabra,palabras_distintas)
    }
    return (D2)
}
```

```r
palabras_dif = palabras_dis(datos)
print(palabras_dif)
```
