## Evaluar la estabilidad del método Borda. Dados n,k,m
 - Construir n candidatos {1,2,...,n}
 - Hacer m copias de los candidatos
 - Aplicar k intercambios aleatorios en cada copia(se vale repetir intercambios)
    - A partir de que valores n,k y m el método ya no puede reconstruir el ranking original {1,2,..,n}

```python
import random
import string
import numba
```

```python
def borda(rankings,pesos):
    assert isinstance(rankings,list)
    assert len(rankings)>1
    assert all(isinstance(ranking,list) for ranking in rankings)
    assert all(isinstance(peso,(int,float)) for peso in pesos)
    elementos = set(rankings[0])
    assert all((elementos==set(ranking)) for ranking in rankings)

    mezcla = dict.fromkeys(rankings[0],0)
    for ranking in rankings:
        for (elemento,peso) in zip(ranking,pesos):
            mezcla[elemento] += peso
    return mezcla
```

```python
@numba.jit
def intercambio_aleatorio(copia):
    r,q=random.randint(0,len(copia)-1),random.randint(0,len(copia)-1)
    if r!=q:
        copia[r],copia[q] = copia[q],copia[r]
    else:
        intercambio_aleatorio(copia)
    return copia
```

```python
def comparar_borda(n,m,k):
    A = list(range(1,n+1))
    copias = [A.copy() for l in range(m)]
    for copia in copias:
        for i in range(k):
            intercambio_aleatorio(copia)
    ranking = borda(copias,A)
    #print(sorted(ranking.keys(),key=ranking.get)!=A)
    return sorted(ranking.keys(),key=ranking.get)!=A
```

```python
def tronar_borda(l,t,y):
    for n in range(2,l+1):
        for m in range(2,t+1):
            comparacion=[]
            for k in range(1,y+1):
                comparacion.append(comparar_borda(n,m,k))
                if comparacion.count(True) > m//.8:
                    print("Truena")
                    break
    return n,m,k
```
