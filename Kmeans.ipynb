```python
k_range = range(1,6)
inertia_list = []
k_list = []
for k in k_range:
    kmeans = cluster.KMeans(n_clusters=k).fit(iris.data)
    #print(knn)
    inertia_list.append(kmeans.inertia_)
```

```python
plt.plot(k_range,inertia_list,marker='.')
plt.xlabel(u'k')
plt.ylabel(u'Inercia')
plt.show()
```

# Medicina

```python
def obtener_valores_numericos(tabla):
    #tabla = pd.read_csv(path)
    numericas = (tabla.dtypes == np.dtype('float64')) | (tabla.dtypes == np.dtype('int64'))
    columnas_numericas = [key for (key, value) in numericas.to_dict().items() if value]
    tabla_num = tabla[columnas_numericas]
    return tabla_num.dropna()
```

```python
df_medicina = pd.read_csv("Kaggle/mexican_medical_students_mental_health_data.csv")
```

```python
medicina = obtener_valores_numericos(df_medicina)
```

```python
kmeans = cluster.KMeans(n_clusters=2).fit(medicina.values)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(30,20))
dendrogram = sch.dendrogram(sch.linkage(iris.data, method = 'ward'))

plt.title('Dendograma')
plt.show()
```
