### Figure 2.6 - Spearman Correlation Coefficient & Pearson Coefficient

```python
np.random.seed(3)
x = np.random.uniform(-3,3,80)
y = np.absolute(x)

rho, pval = stats.spearmanr(x, y)
pearson = stats.pearsonr(x, y)

title = "Spearman coefficient = " + str("%.2f" % rho) + "\nPearson coefficient = " + str("%.2f" % pearson[0])
plt.title(title)

plt.plot(x,y,'ro')
```

### Figure 2.7 - Spearman Correlation Coefficient & Pearson Coefficient ( 3 graphs)

```python
plt.figure(figsize=(12,4.3))

plt.subplot(131)
x0 = np.linspace(1,5,20)
y0 = 0.1*x0
x = np.linspace(5.04,5.4,25)
y = 13*(x-5)

xval = np.concatenate([x0,x])
yval = np.concatenate([y0,y])
rho, pval = stats.spearmanr(xval, yval)
pearson = stats.pearsonr(xval, yval)

m, b = np.polyfit(xval, yval, 1)
rline_xval = np.concatenate([xval, [0,6]])
plt.plot(xval, yval, 'yo', rline_xval, m*rline_xval + b, 'k-')
title = "Spearman coefficient = " + str(rho) + "\nPearson coefficient = " + str("%.2f" % pearson[0])
plt.title(title)
plt.ylim(0,6)
plt.xlim(0,6)

plt.subplot(132)
np.random.seed(3)
x = np.random.uniform(0.05,0.95,75)
noise = np.random.normal(0,0.1,75)
y = x + noise
rho, pval = stats.spearmanr(x, y)
pearson = stats.pearsonr(x, y)
m, b = np.polyfit(x, y, 1)
rline_xval = np.concatenate([x, [0,1]])

plt.plot(x, y, 'yo', rline_xval, m*rline_xval + b, 'k-')
title = "Spearman coefficient = " + str("%.2f" % rho) + "\nPearson coefficient = " + str("%.2f" % pearson[0])
plt.title(title)
plt.xlim(0,1)
plt.ylim(-0.2,1.2)

plt.subplot(133)
np.random.seed(3)
x = np.random.uniform(0.05,0.95,75)
noise = np.random.normal(0,0.1,75)
y = x + noise
y[29] += 4

rho, pval = stats.spearmanr(x, y)
pearson = stats.pearsonr(x, y)
m, b = np.polyfit(x, y, 1)
rline_xval = np.concatenate([x, [0,1]])

plt.plot(x, y, 'yo', rline_xval, m*rline_xval + b, 'k-')
title = "Spearman coefficient = " + str("%.2f" % rho) + "\nPearson coefficient = " + str("%.2f" % pearson[0])
plt.title(title)
plt.xlim(0,1)
plt.ylim(-0.2,5.2)

plt.tight_layout() #Reduce the overlap of labels
```

### Figure 2.9

```python
fig = plt.figure(figsize=(10,5))

# Making Data Points
np.random.seed(3)
x = np.random.uniform(0.05,0.95,75)
noise = np.random.normal(0,0.1,75)
y = x + noise

meany = sum(y)/len(y)
y = [i-meany for i in y]
rho, pval = stats.spearmanr(x, y)
pearson = stats.pearsonr(x, y)

print( "mean ", meany)
print("var = ", np.var(y)) 
print("Pearson = ", pearson[0])

plt.subplot() 

# LEFT PLOT(121)
plt.plot(x, y, 'ro')

m, b = np.polyfit(x, y, 1)
rline_xval = np.concatenate([x, [0,1]])
plt.plot(rline_xval, m*rline_xval + b, 'r-', linewidth=1.2)

title = "Spearman coefficient = " + str(rho) + "\nPearson coefficient = " + str(pearson[0])
#plt.title(title)
plt.xlim(0,1)
plt.ylim(-0.6,0.6)
plt.xlabel("x")
plt.ylabel("y")

# RIGHT PLOT
plt.subplot(122)
regline = [m*i + b for i in x]
k =[i - j for i,j in zip(y, regline)]
plt.plot(x, k, 'ro')

m, b = np.polyfit(x, k, 1)
rline_xval = np.concatenate([x, [0,1]])
plt.plot(rline_xval, m*rline_xval + b, 'r-', linewidth=1.2)

plt.ylim(-0.6, 0.6)
plt.xlabel("x")
plt.ylabel("y - f(x)")

plt.tight_layout()
pearson = stats.pearsonr(x, k)
print( "var2 = ", np.var(k))
print( "Pearson2 = ", pearson[0])
```
