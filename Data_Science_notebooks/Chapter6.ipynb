### Figure 6.10 Color bars

```python
plt.figure(figsize=(10,4.5))
gradient = np.linspace(0, 1, 256)
# Stack arrays in sequence vertically (row wise).
gradient = np.vstack((gradient, gradient)) 

x = 7
i = 1

ax = plt.subplot(x, 1, i)
name = 'Greens'
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("Greens", rotation='horizontal', labelpad=40)
i +=1

ax1 = plt.subplot(x, 1, i)
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap('gray'))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("gray", rotation='horizontal', labelpad=40)
i +=1

ax2 = plt.subplot(x, 1, i)
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap('cool'))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("cool", rotation='horizontal', labelpad=40)
i +=1

ax3 = plt.subplot(x, 1, i)
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap('spring'))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("spring", rotation='horizontal', labelpad=40)
i +=1

ax4 = plt.subplot(x, 1, i)
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap('seismic'))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("seismic", rotation='horizontal', labelpad=40)
i +=1

ax5 = plt.subplot(x, 1, i)
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap('rainbow'))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("rainbow", rotation='horizontal', labelpad=40)
i +=1

ax6 = plt.subplot(x, 1, i)
plt.imshow(gradient, aspect='auto', cmap=plt.get_cmap('gist_rainbow'))
plt.tick_params(
    axis='both', which='both', bottom='off', top='off',         
    left='off', right='off', labelbottom='off', labelleft='off') 
plt.ylabel("gist_rainbow", rotation='horizontal', labelpad=40)
```

### Figure 6.15 Data

```python
with xport.XportReader("/home/yeseul/notebook/BMX_H.XPT") as reader:
        Z = [row for row in reader]
Ix = [z.pop("SEQN") for z in Z]
A = pd.DataFrame(Z, index=Ix,columns=['BMXBMI','BMXHT','BMXWT'])

A['ht_m']=(A['BMXHT'])*0.01

normal=A[(A['BMXBMI']>=18.5) & (A['BMXBMI']<=24.9)]
normal=pd.DataFrame(normal)

overweight=A[(A['BMXBMI']>=25) & (A['BMXBMI']<=29.9)]
overweight=pd.DataFrame(overweight)

obese=A[(A['BMXBMI']>=30)]
obese=pd.DataFrame(obese)
```

### Figure 6.15 Left

```python
%matplotlib inline
import matplotlib.pyplot as plt
opacity = 0.05

ax=normal.plot(kind='scatter', x='BMXWT', y='ht_m',c='b',figsize=(7,4),alpha=opacity, linewidth='0.0')         
overweight.plot(kind='scatter',x='BMXWT',y='ht_m',ax=ax,c='b',alpha=opacity, linewidth='0.0')
obese.plot(kind='scatter',x='BMXWT',y='ht_m',ax=ax,c='b',alpha=opacity, linewidth='0.0')
ax.set_xlabel('WEIGHT (Kgs)')
ax.set_ylabel('HEIGHT (Metres)')
ax.set_title('Distribution of Americans On The Basis Of BMI')

plt.xlim(50,130)
plt.ylim(1.3,2.0)
```

### Figure 6.15 Right

```python
#Heat map
x = np.append(normal['BMXWT'], obese['BMXWT'])
x = np.append(x, overweight['BMXWT'])
y = np.append(normal['ht_m'], obese['ht_m'])
y = np.append(y, overweight['ht_m'])
xy = np.vstack([x,y])
z = stats.gaussian_kde(xy)(xy)
# Reference: http://stackoverflow.com/questions/20105364/how-can-i-make-a-scatter-plot-colored-by-density-in-matplotlib

#continued heat map
plt.scatter(x, y, c=z, edgecolor='', s = 7)
plt.xlabel('WEIGHT (Kgs)')
plt.ylabel('HEIGHT (Metres)')
plt.title('Distribution of Americans by Height and Weight')
plt.xlim(50,130)
plt.ylim(1.3,2.0)
```

### Figure 6.23 
Frequency diagram with normal distribution.

```python
mu = 100
sigma = 1 
N = 100

# Draw random samples from a normal (Gaussian) distribution.
np.random.seed(3)
s = np.random.normal(mu, sigma, N)
plt.figure(figsize=(12,4.5))

#Draw the histogram
plt.subplot(121)
n, bins, patcheds = plt.hist(s, 30, color='red')
plt.xlim(97,102.2)
plt.ylim(0,10)
plt.title("Frequency Histogram")
width = (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2

#Draw the density histogram (bar graph)
plt.subplot(122)
plt.bar(center, n/sum(n), align='center', width = width, color='red')
plt.xlim(97,102.2)
plt.ylim(0.0, 0.10)
plt.title("Density Histogram")
```

### Data for Figure 6.29
#### Data Source: NYC Weather Data from the link below
https://www.wunderground.com/history/airport/KNYC/2015/1/1/CustomHistory.html?dayend=31&monthend=12&yearend=2015&req_city=&req_state=&req_statename=&reqdb.zip=&reqdb.magic=&reqdb.wmo=&MR=1

```python
path = "/home/yeseul/practice-notebook/weather.xlsx"
data = pd.ExcelFile(path).parse('Sheet1')

avghigh = [int(re.search(r'[-+]?\d+', data["AvgHigh"][i][:2]).group()) for i in range(365)]
avglow = [int(re.search(r'[-+]?\d+', data["AvgLow"][i][:2]).group()) for i in range(365)]
rechigh = [int(re.search(r'[-+]?\d+', data["RecHigh"][i][:3]).group()) for i in range(365)]
reclow = [int(re.search(r'[-+]?\d+', data["RecLow"][i][:3]).group()) for i in range(365)]

breakhigh = []
breaklow = []
for i in range(365):
    if data["Max TemperatureF"][i] > rechigh[i]:
        breakhigh.append(i)
    if data["Min TemperatureF"][i] < reclow[i]:
        breaklow.append(i)
```

### Figure 6.29

```python
plt.figure(figsize=(12,6))
x = range(365)

plt.fill_between(x, avglow, avghigh, facecolor='orange', alpha=0.8,\
                label="Average Range")
plt.fill_between(x, reclow, rechigh, facecolor='orange', alpha=0.6,\
                label="Record Range")
plt.fill_between(x, data["Min TemperatureF"], data["Max TemperatureF"], facecolor='blue', alpha=0.50,\
                label = "Actual Range")

#Breakthroughs
plt.plot(breakhigh, data["Max TemperatureF"][breakhigh], 'ro', label="New Record High")
plt.plot(breaklow, data["Min TemperatureF"][breaklow], 'bo', label="New Record Low")
plt.xlim(0,365)
plt.ylim(-10,110)

plt.xticks([0,31,59,90,120,151,181,212,243,273,304,334,365,366], \
           ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec", "Jan"])
plt.ylabel("Temperature (F)")
plt.title("New York City Weather 2015", fontsize=16)
plt.legend(loc="lower center")
```
